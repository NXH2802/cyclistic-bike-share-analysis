```python
# thư viện oprating system: thao tác với hệ điều hành, đổi tập tin..
import os
```

* cách dùng concat gộp dataframe
* frames = [df1, df2, df3]

```python
frames = []
# filepaths = [] bỏ do không cần đường dẫn nữa
for file in os.listdir(path):
    if file.endswith("csv"):
        filepath = path + file
        # filepaths.append(filepath): không cần  
        df1 = pd.read_csv(filepath)
        frames.append(df1)
# print(filepaths)
result = pd.concat(frames)
result
```

```python
# Kiểm tra độ dài từng tập tin
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith("csv"):
        filepath = path + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)
```

```python
print(all_length)
```

```python
# In ra độ dài tập tin khi sau khi nối
print(sum(all_length))
```

```python
df = result
```

```python
# Truyển kiều dữ liệu của thời gian bắt đầu kết thúc -> sang định dạng ngày tháng để phục vụ tính toán
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
df['ride_length'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

df.head()
```

```python
# Kiểm tra số chuyến có thời gian âm
count_negative = df[df['ride_length'] < 0].shape[0]
print(f"Số chuyến có thời gian âm: {count_negative}")

# Kiểm tra số chuyến có thời gian lớn hơn 24 giờ (1440 phút)
count_over_24h = df[df['ride_length'] > 1440].shape[0]
print(f"Số chuyến có thời gian trên 24 giờ: {count_over_24h}")
```
